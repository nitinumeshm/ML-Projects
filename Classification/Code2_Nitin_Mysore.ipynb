## Business Understanding


Bank wants to make use of a classification model which can predict whether the client has subscribed to a term deposit. The purpose of this classification model is to identify in the future whether a client will subscribe to a term deposit or not.

As per the business requirement, the evaluation metrics will be more focused on Recall and Precision, rather than Accuracy. Recall means the model will find those customers who are more likely to subscribe to a term deposit, and Precision means out of those customers, who will actually subscribe.

Both of these metrics will aid in optimising the bank's resources in marketing for term deposits.

# *Part One*

*In this part, we will use all the features to train the machine learning models.*

## Data Preparation

```python
## descriptive statistics of the numerical data
data.describe()
```

### Data Cleaning

As there are no missing values in the dataset, skipping this step.

### Data Encoding



*   For features that can be ranked, we will use *map()* method to replace feature values.
*   For features that cannot be ranked, we will use *get_dummies()* method to replace feature values.

The purpose of data encoding is to convert categorical data into numerical data so that the machine learning algorithms can process and interpret the data.



```python
"""Ranked Features"""

## Data encoding for education
data['education'] = data['education'].map({'unknown': 0, 'primary': 1, 'secondary': 2, 'tertiary': 3})

## Data encoding for default
data['default'] = data['default'].map({'no': 0, 'yes': 1})

## Data encoding for housing
data['housing'] = data['housing'].map({'no': 0, 'yes': 1})

## Data encoding for loan
data['loan'] = data['loan'].map({'no': 0, 'yes': 1})

## Data encoding for month
data['month'] = data['month'].map({'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6, 'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12})

## Data encoding for y (target variable)
data['y'] = data['y'].map({'no': 0, 'yes': 1})
```

```python
"""Un-ranked features"""

from pandas import get_dummies

data = get_dummies(data, columns = ['job', 'marital', 'contact', 'poutcome'], dtype = int)
```

```python
## check if data has been transformed or not
data.info()
```

Based on the output above, the data has now a total of 35 columns. This means that the dataset has been encoded properly.

```python
## save the data into a csv file
data.to_csv('bank_encoded.csv')
```

### Divide the dataset into x (features) and y (target variable)

```python
## divde the data
y = data['y']
x = data.drop('y', axis = 1)

print("Shape of x: ", x.shape)
print("Shape of y: ", y.shape)
```

### Data Scaling (only for x)

In this step, we will transform the data into a standard normal distribution using StandardScaler. The purpose of data scaling is to transform the data prepared till now into a format that is suitable for the machine learning model.

```python
## transform the data into standard normal distribution
from sklearn.preprocessing import StandardScaler
from pandas import DataFrame

x_scaled = StandardScaler().fit_transform(x)
DataFrame(x_scaled).to_csv('bank_scaled.csv', index = False)
```

### Data Splitting

Divide the data into training set, and test set. The purpose of the training set is to make the model learn. Then using the test set, we shall see how well the model learned.

```python
## split the data into training and test set
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.25, random_state = 10)

print("x train set: ", x_train.shape)
print("y train set: ", x_test.shape)

print("\nx test set: ", x_test.shape)
print("y test set: ", y_test.shape)
```

### Data Balancing

Using SMOTE, we will create some synthetic data in order to balance the output training set *(y_train)*. The purpose of data balancing step in data preparation is to fix any imbalances in the dataset so as to prevent machine learning models being biased towards classes that have a majority presence.

```python
## Check if y_train has equal number of 0s and 1s
y_train.value_counts()
```

```python
## values are imbalanced, so balance it using SMOTE

from imblearn.over_sampling import SMOTE

x_train, y_train = SMOTE().fit_resample(x_train, y_train)

y_train.value_counts()
```

Now, the dataset is balanced and ready for the next phase, which is Modeling.

## Modeling, Evaluation and Deployment

### Decision Tree Classifier

A Decision Tree Classifier is a supervised ML algorithm that categorizes data by recursively splitting it into smaller sets, till it can surely decide a label.

```python
from sklearn.tree import DecisionTreeClassifier
```

#### Method 1

##### Modelling

```python
## build the model with max_depth manually entered
dt_classifier = DecisionTreeClassifier(max_depth = 3)
```

To avoid underfitting or overfitting, train the model using a range of *max_depth* values, later to find a value which makes the model generalised.

```python
## train the model using x_train and y_train dataset
dt_classifier.fit(x_train, y_train)
```

```python
## test the model using x_test
y_pred_dt = dt_classifier.predict(x_test)
y_pred_dt
```

##### Evaluate the model

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

## Confusion Matrix
conf_matrix_dt = confusion_matrix(y_test, y_pred_dt)
print("Confusion Matrix: \n", conf_matrix_dt)

## Accuracy
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print("\nAccuracy: ", round(accuracy_dt * 100, 2), "%")

## Recall
recall_dt = recall_score(y_test, y_pred_dt)
print("\nRecall: ", round(recall_dt * 100, 2), "%")

## Precision
precision_dt = precision_score(y_test, y_pred_dt)
print("\nPrecision: ", round(precision_dt * 100, 2), "%")

## F1 Score
f1_dt = f1_score(y_test, y_pred_dt)
print("\nF1 Score: ", round(f1_dt * 100, 2), "%")

## deployment decision
if recall_dt > 70 and precision_dt > 70:
  print("\nModel recommended!")
else:
  print("\nModel not recommended!")
```

#### Method 2

```python
from sklearn.model_selection import GridSearchCV
```

```python
## define the estimator
dt_classifier_estimator = DecisionTreeClassifier(criterion='entropy')

## define the depth
dt_classifier_depths = {
    'max_depth': [2, 3, 5, 7, 11, 14, 16, 20, 25, 30]
}

## build the model
grid_search_dt_classifier = GridSearchCV(estimator = dt_classifier_estimator, param_grid = dt_classifier_depths, scoring = 'recall', cv = 5)
```

```python
## train, test, evaluate, and rank
grid_search_dt_classifier.fit(x_scaled, y)
```

```python
## best max depth
best_max_depth_dt = grid_search_dt_classifier.best_params_['max_depth']
print("Best Max Depth: ", best_max_depth_dt)

## best recall score
best_recall_dt = grid_search_dt_classifier.best_score_
print("\nBest Recall Score: ", round(best_recall_dt * 100, 2), "%")

## best decision tree
best_dt = grid_search_dt_classifier.best_estimator_
print("\nBest Decision Tree: \n", best_dt)

## deployment decision
if best_recall_dt > 70:
  print("\nModel recommended!")
else:
  print("\nModel not recommended!")
```

### Random Forest Classifier

In Random Forest Classifier, instead of building a single decision tree and use it to make predictions, build many slightly different trees and combine their
predictions.

Random Forest Classifier is a type of Ensemble Learning that combines multiple learning algorithms to obtain performance improvements over its components.

```python
from sklearn.ensemble import RandomForestClassifier
```

#### Method 1

##### Modelling

```python
## build the model
rf_classifier = RandomForestClassifier(criterion = 'entropy', n_estimators = 100)

## train the model using x_train and y_train datasets
rf_classifier.fit(x_train, y_train)

## test the model
y_pred_rf = rf_classifier.predict(x_test)
y_pred_rf
```

To avoid overfitting or underfitting, train the model with a range of *n_estimators* (number of trees), to find the value that generalises the model.

##### Evaluate the model

```python
## Confusion Matrix
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
print("Confusion Matrix: \n", conf_matrix_rf)

## Accuracy
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("\nAccuracy: ", round(accuracy_rf * 100, 2), "%")

## Recall
recall_rf = recall_score(y_test, y_pred_rf)
print("\nRecall: ", round(recall_rf * 100, 2), "%")

## Precision
precision_rf = precision_score(y_test, y_pred_rf)
print("\nPrecision: ", round(precision_rf * 100, 2), "%")

## F1 Score
f1_rf = f1_score(y_test, y_pred_rf)
print("\nF1 Score: ", round(f1_rf * 100, 2), "%")

## decision support
if recall_rf > 70 and precision_rf > 70:
  print("\nModel recommended!")
else:
  print("\nModel not recommended!")
```
