```python
""" prepare the data for prophet """

## prophet requires specific column names "ds" and "y"
prophet_df = euro_to_inr.reset_index()[["date", "eur_inr"]]
prophet_df.columns = ["ds", "y"]

## inspect the prophet dataframe
prophet_df.head()
```

```python
""" initialise the prophet """

model = Prophet(
    daily_seasonality = False,
    weekly_seasonality = False,
    yearly_seasonality = True
)
```

```python
""" train the model """

model.fit(prophet_df)
```

```python
""" create future dates (next 30 days) """

future = model.make_future_dataframe(periods=30)

## check
future
```

```python
""" generate forecast """

forecast = model.predict(future)
forecast
```

```python
""" plot the forecast """

model.plot(forecast)
```

```python
""" plot components, it shows trends + seasonality """

model.plot_components(forecast)
```
