Data pre-processing for time-series modeling

```python
""" load the data """

df.to_csv("euro_to_inr.csv", index=False)
euro_to_inr = pd.read_csv("euro_to_inr.csv")
euro_to_inr["date"] = pd.to_datetime(df["date"])
euro_to_inr = euro_to_inr.sort_values("date").reset_index(drop=True)
```

```python
""" set date as time index as the time-series model expects a proper date index """

euro_to_inr = euro_to_inr.set_index("date")
```

```python
""" check for missing dates, as APIs usually skip weekends and holidays which creates gaps """
full_range = pd.date_range(
    start = euro_to_inr.index.min(),
    end = euro_to_inr.index.max(),
    freq = "D"
)

missing_dates = full_range.difference(df.index)
print(f"Missing dates: {len(missing_dates)}")
```

```python
""" add missing dates to the dataframe by re-indexing """

euro_to_inr = euro_to_inr.reindex(full_range)
euro_to_inr.index.name = "date"
```

```python
""" Handle missing values,
as the rate usually remains the same over the weekends,
most realistic approach is to forward-fill """

euro_to_inr['eur_inr'] = euro_to_inr['eur_inr'].ffill()
print(euro_to_inr.isna().sum())
```

```python
""" visualise the cleaned data """

plt.figure(figsize=(12, 6))
plt.plot(euro_to_inr.index, euro_to_inr["eur_inr"])
plt.title("Euro to INR Exchange Rate (Daily and Cleaned)")
plt.xlabel("Date")
plt.ylabel("INR value for 1 Euro")
plt.show()
```
