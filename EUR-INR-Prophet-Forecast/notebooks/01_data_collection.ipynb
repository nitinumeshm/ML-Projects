I'm choosing Frankfurter Exhange Rates API as it is completely free, and there's no API key required. This API provides a reliable historical data.

Link: https://frankfurter.dev

If you paste the link in the browser, the output you get is a JSON format as below

```
{
  "amount":1.0,
  "base":"EUR",
  "date":"2026-01-22",
  "rates":
    {
        "INR":107.216
    }
}
```

Fetch historical data of EURO > INR using Python and convert it into a clean time-series DataFrame.

````python
## import necessary libraries
import requests
import pandas as pd
import matplotlib.pyplot as plt
from datetime import date
````

````python
""" decide the starting time of the data that you want to collect """

## start date from January 2023
start_date = "2023-01-01"

## till present day
end_date = date.today().isoformat()

## connect with the API
api_url = f"https://api.frankfurter.dev/v1/{start_date}..{end_date}"

## fetch the data from the API
params = {
    "base": "EUR",
    "symbols": "INR"
}

response = requests.get(api_url, params=params)
data = response.json()

## convert json to dataframe format
rates = data["rates"]
df = pd.DataFrame([
    {"date": d, "eur_inr": rates[d]["INR"]}
    for d in rates
])

## convert "date" into proper datetime format
df["date"] = pd.to_datetime(df["date"])
````

````python
""" plot the data """

## lets plot the graph and see the trend
plt.figure(figsize=(12, 6))
plt.plot(df["date"], df["eur_inr"])
plt.title("Euro to INR Exchange Rate")
plt.xlabel("Date")
plt.ylabel("INR value for 1 Euro")
plt.show()
````
